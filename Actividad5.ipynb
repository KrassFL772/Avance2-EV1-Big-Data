```python
import math;

base = 4
altura = 6

area_triangulo = base * altura
print("------Calcular área de un triángulo.------\n")
print("El área del triángulo es: ", area_triangulo, "\n")

"""Determinar la hipotenusa de un triángulo."""

from math import sqrt

try:
    input = raw_input
except NameError:
    pass
def get_numeric_value(prompt):

    while True:
        value = input(prompt)
        if not value:
            return None
        try:
            value = float(value)
        except ValueError:
            print("Debe ingresar un valor numerico. "
                  "Vuelva a intentarlo.")
        else:
            break
    return value

def get_triangle(a=None, b=None, h=None):

    print("------Obtener la hipotenusa de un triángulo.------\n")

    return r"""
    |\
    | \
    |  \ h
  a |   \
    |    \
    |_ _ _\
      b

    a = %s
    b = %s
    h = %s""" % ("?" if a is None else a,
                 "?" if b is None else b,
                 "?" if h is None else h)


def semiperimetro(a, b, h):
  semi = (a + b + h) / 2

  return semi

"""Calcular seno, coseno, tangente de un triángulo."""

def sencostan_trgl(x, y):
    print("------Calcular seno, coseno, tangente de un triángulo.------\n")
    # Calcular la hipotenusa utilizando el teorema de Pitágoras
    h = math.sqrt(x**2 + y**2)

    # Calcular el seno, coseno y tangente
    seno = y / h
    coseno = x / h
    tangente = y / x  # No es necesario calcular la hipotenusa para obtener la tangente

    # Convertir los valores a grados
    seno_grados = math.degrees(math.asin(seno))
    coseno_grados = math.degrees(math.acos(coseno))
    tangente_grados = math.degrees(math.atan(tangente))


    return seno_grados, coseno_grados, tangente_grados

# Llamar a la función con los valores X = 12.3 y Y = 11
resultado_seno, resultado_coseno, resultado_tangente = sencostan_trgl(12.3, 11)

print("El seno del ángulo es: ", resultado_seno)
print("El coseno del ángulo es: ", resultado_coseno)
print("La tangente del ángulo es: ", resultado_tangente, "\n")

def main():
    print(get_triangle())
    print("\nDeje en blanco el valor que desconoce.\n")

    a = get_numeric_value("a: ")
    b = get_numeric_value("b: ")

    if a is None or b is None:
        h = get_numeric_value("h: ")
    else:
        h = None

    if (bool(a) + bool(b) + bool(h)) < 2:
        print("Debe especificar, al menos, dos valores.")
        return 0

    if h is None:
        h = sqrt(a**2 + b**2)
    elif a is None:
        a = sqrt(h**2 - b**2)
    elif b is None:
        b = sqrt(h**2 - a**2)

    print("El semiperímetro es: ",semiperimetro(a, b, h))



    print("")
    print(get_triangle(a, b, h))

    return 0
if __name__ == '__main__':
    main()

print("\n\n\n\n\n")
print("Developed by: Ángel Gabriel Hernández Flores - Universidad TecMilenio.")

```
