```python
import pandas as leer

tarifas = leer.read_csv('Tarifas por zonas 2016-2017.csv')

print(tarifas)

#Validar costo promedio tarifas por zonas.
tarifas_volumetricro = tarifas['volumetrica']
print("Costo promedio tarifas por zonas: \n",tarifas_volumetricro, "\n")

#Función que calcule la variación del precio del gas natural a lo largo de dos años.
tarifas['fecha_inicio'] = leer.to_datetime(tarifas['fecha_inicio'])
print("Variación del precio del gas natural a lo largo de dos años: \n",tarifas, "\n")
año_2016 = tarifas[tarifas['fecha_inicio'].dt.year == 2016]
print(año_2016, "\n")

#Demanda promedio del gas natural para 2016.
print("Demanda promedio del gas natural para 2016: ",año_2016['capacidad_base_firme'].sum(), ", Longitud: ", len(año_2016),"\n")

#Demanda promedio del gas natural para 2016.
año_2017 = tarifas[tarifas['fecha_inicio'].dt.year == 2017]
print("Demanda promedio del gas natural para 2017: ",año_2017['capacidad_base_firme'].sum(), ", Longitud: ", len(año_2017),"\n")

#Función que devuelva las tarifas máximas para cada uno de los dos años.
tarifa_max =[(año_2016['maxima_base_interrumpible'] >= 1) & (tarifas['fecha_inicio'].dt.year == 2016)]
print("Tarifas máximas para el año 2016: ",año_2016['maxima_base_interrumpible'].sum(), ", Longitud: ", len(año_2016),"\n")
tarifa_max =[(año_2017['maxima_base_interrumpible'] >= 1) & (tarifas['fecha_inicio'].dt.year == 2017)]
print("Tarifas máximas para el año 2017: ",año_2017['maxima_base_interrumpible'].sum(), ", Longitud: ", len(año_2017),"\n")


```
